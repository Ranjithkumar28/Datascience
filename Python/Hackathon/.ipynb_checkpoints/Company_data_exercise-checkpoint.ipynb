## Analysing the given data

```python
#Import Library
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer
import matplotlib.pyplot as plt 
%matplotlib inline 
import seaborn as sns
```

```python
df = pd.read_csv('company.csv')
```

```python
df.drop_duplicates(inplace = True)
```

```python
df.shape
```

```python
df
```

## Handling the null colums in the data

```python
df['Company'].unique()
```

```python
df['Company'] = df['Company'].replace(['Tata Consultancy Services', 'Congnizant', 'Infosys Pvt Lmt'], ['TCS', 'CTS', 'Infosys'])
```

```python
df
```

```python
df['Company'].unique()
```

```python
df['Company'] = df['Company'].fillna('Others')
```

```python
df['Company'].unique()
```

```python
df['Place'] = df['Place'].fillna('Others')
df['Place'].unique()
```

## Replace the mean for Age and Salary using the preprocessing techniques

```python
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
imputer = imputer.fit(df[['Age']])
df['Age'] = imputer.transform(df[['Age']])
```

```python
df.sample(50)
```

```python
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
imputer = imputer.fit(df[['Salary']])
df['Salary'] = imputer.transform(df[['Salary']])
```

```python
df
```

```python
df.isna().sum()
```

```python
df['Gender'] = df.Gender.map({0:'F', 1:'M'})
```

```python
df
```

## Visualization of data

# Count of Employees of Each Company

## TCS has highest number of employees

```python
# setting up the size of  charts
plt.title("Count of Employees by Company wise")

#defining bar plot here
df['Company'].value_counts().plot(kind= 'pie', autopct='%1.2f', figsize = (16,6), startangle=90)
plt.show()
```

## Salary based on the city

```python
# setting up the size of  charts
plt.figure(figsize = (16,6))
plt.title("Sum of Salary by Place")

#defining bar plot here
sns.barplot(x = df['Place'],  y = df['Salary'])

#Adding labels
plt.xlabel("State")
plt.ylabel("Count")

plt.show()
```

## Comparison of Age and Salary

```python
# setting up the size of  charts
plt.figure(figsize = (16,6))
plt.title("Comparison of Age and Salary")

#defining bar plot here
sns.barplot(x = df['Age'],  y = df['Salary'])

#Adding labels
plt.xlabel("Salary")
plt.ylabel("Age")

plt.show()
```

# Female employees are more than male

```python
# setting up the size of  charts
plt.title("Count of Employees by Company wise")

#defining bar plot here
df['Gender'].value_counts().plot(kind= 'pie', autopct='%1.2f', figsize = (16,6), startangle=90)
plt.show()
```

## Box plot for Age, Salary and Gender

```python
df.plot.box(figsize=(10,8))
plt.show()
```

## Pair plot for Age, Salary and Gender

```python
g = sns.pairplot(df)
g.set(xticklabels = [])
plt.show()
```

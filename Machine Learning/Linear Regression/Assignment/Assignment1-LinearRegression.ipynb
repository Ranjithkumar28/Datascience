# Problem Statement
#### You just got some contract work with an Ecommerce company based in New York City that sells clothing online but they also have in-store style and clothing advice sessions. Customers come in to the store, have sessions/meetings with a personal stylist, then they can go home and order either on a mobile app or website for the clothes they want
#### The company is trying to decide whether to focus their efforts on their mobile app experience or their website. They've hired you on contract to help them figure it out.

```python
# Import the library
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression, Lasso, Ridge
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
# Reading the data from csv file
df = pd.read_csv('Ecommerce_Customers.csv')
```

```python
df.head()
```

### Checking if any missing values in the data

```python
df.info()
```

```python
sns.pairplot(df)
```

### If you look at the above plot, when compare to other independent variables only Length of Membership is highly correlated with the Yearly Amount Spent. See the correlation differences in the below plot:

```python
plt.figure(figsize=(15,10))
sns.heatmap(df.corr(), annot=True)
```

### Test and Train Split for LR

```python
X = df.drop(['Email', 'Avatar', 'Address', 'Yearly Amount Spent'],axis=1)
y = df['Yearly Amount Spent']
```

```python
X_train, X_test, y_train, y_test = train_test_split(
     X, y, test_size=0.4, random_state=101)
```

```python
X_train.shape
```

```python
X_test.shape
```

### Fitting test data into the LR Model

```python
lr = LinearRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
print("Coefficients are : {}".format(lr.coef_))
```

```python
print("Coefficients is : {}".format(lr.intercept_))
```

### Predicting test data

```python
predictions = lr.predict(X_test)
```

```python
plt.figure(dpi = 150)
plt.scatter(y_test, predictions)
plt.plot(y_test, y_test, color="black")
plt.xlabel("Actual values")
plt.ylabel("Predicted values")
plt.show()
```

### Evaluating the Model with Error Metrics

```python
print('Mean Absolute Error:',mean_absolute_error(y_test,predictions))
print('Mean Squared Error:',mean_squared_error(y_test,predictions))
print('Root Mean Absolute Error:',np.sqrt(mean_squared_error(y_test,predictions)))
print('R2 Square:',r2_score(y_test,predictions))
```

### Normalizing the data

```python
from sklearn.preprocessing import StandardScaler
```

```python
scx = StandardScaler()
```

```python
X_train_std = scx.fit_transform(X_train)
```

```python
X_test_std = scx.transform(X_test)
```

### Using Ridge

```python
for i in [10,100, 1000, 10000, 100000, 1000000, 10000000]:
    ridge = Ridge(alpha=i)
    ridge.fit(X_train_std, y_train)
    pred_ridge = ridge.predict(X_test_std)
    print("Lambda is "+str(i)+" ", ridge.coef_)
    print(mean_absolute_error(y_test,pred_ridge))
    print(r2_score(y_test,pred_ridge))
```

### Using Lasso

```python
for i in [10,100, 1000, 10000, 100000, 1000000, 10000000]:
    lasso = Lasso(alpha=i)
    lasso.fit(X_train_std, y_train)
    pred_lasso = lasso.predict(X_test_std)
    print("Lambda is "+str(i)+" ", lasso.coef_)
    print(mean_absolute_error(y_test,pred_lasso))
    print(r2_score(y_test,pred_lasso))
```
